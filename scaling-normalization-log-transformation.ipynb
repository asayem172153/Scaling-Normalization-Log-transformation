# Standardization

```python
df3=df3.drop('Profit',axis=1)
```

```python
df3.head()
```

```python
import math
def sd(column):
    tsum = 0
    mean = column.mean()
    column=column.fillna(mean)
    n = len(column.dropna())

    for i in column:
        tsum += (i-mean)**2

    std_cal= math.sqrt(tsum/(n-1))
    return std_cal,mean
```

```python
df3.head()
```

```python
for i in df3.columns:
    if df3[i].dtype != 'object':
        std,mean=sd(df3[i])
        df3[i]=(df3[i]-mean)/std
        
```

# Robust Scaler

```python
df5=df5.drop('Profit',axis=1)
```

```python
df5.head()
```

```python
def IQR(column):
    data=column
    data=sorted(data)
    n= len(data)
    index75 = int(0.75*(n-1))
    index25 = int(0.25*(n-1))
    
    x75 = data[index75]
    x25 = data[index25]
    
    return x75,x25
```

```python
for column in df5.columns:
    if df5[column].dtype != 'object':
        x75,x25 = IQR(df5[column])
        #print('x75: ',x75,'x25: ',x25)
        df5[column] = (df5[column]-df5[column].median())/(x75-x25)
        
```
